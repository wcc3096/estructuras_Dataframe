## M1 T02: Exploración y estructura de los datos - Dataframes con Pandas


Cargamos el archivo CSV.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#cargamos el data set a un dataframe de pandas
df=pd.read_csv(r'C:\Users\wcc30\Desktop\DelayedFlights.csv')

#info
df.info()
```

Hemos decidido primero comenzar con el calculo de datos missing para tener mayor criterio y una causalidad para eliminar variables innecesarias que podrían afectar por su gran cantidad de datos missing.

```python
#Calculamos la suma de los datos faltantes de los datos missing de cada variable:
print("Suma de datos missing:\n", df.isnull().sum())

#Prefiero mejor visualizar los datos missings en %, ya que nos permite tener un mejor análisis de ellos
print("\nPorcentaje de datos missing % :\n", round((df.isnull().sum()*100/df.shape[0]),2))
```

Podemos observar que "CarrierDelay", "WeatherDealy", "NASDelay", "SecurityDelay", "LateAircraftDelay" son las variables que tienen un porcentaje altos de datos missing, por lo cual seran las primeros que eliminaremos. 

Otras variables que decidimos eliminar son: 

-"Cancelled", debido a que hay otra variable similar codificada "CancellationCode".

-"DayOfWeek", dias de la semana son datos irrelevantes.

-"Unnamed: 0"

-"TaxiIn", datos irrelevantes para posterior análisis.

-"TaxiOut", datos irrelevantes para posterior análisis.

Las demás variables que tienen NAs, debido a que es minimo el porcentaje de datos faltantes que tienen, realizaremos la eliminacion de estos datos.

```python
#Eliminamos los NAs restantes
df= df.dropna()

#Eliminamos los registros duplicados
df = df.drop_duplicates()
```

Agruparemos 3 variables: "Year" , "Month" y "DayofMonth" en una sola variable "Data", para posteiormente eliminar las 3 primeras mencionadas

```python
df['Data'] = df['DayofMonth'].map(str) + '-' + df['Month'].map(str) + '-' + df['Year'].map(str)
df.sort_values(['Data'])
```

Eliminamos las variables

```python
df = df.drop(['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay',
         'LateAircraftDelay', 'Unnamed: 0', 'DayOfWeek', 'Cancelled'
        , 'TaxiIn', 'TaxiOut','Year', 'Month', 'DayofMonth'], axis=1)

df.info()
```

Como pudimos ver inicialmente en la info del dataframe,las variables "Diverted" y "FlightNum" estan reconocidas como variables numéricas por lo cual las definimos como categóricas para no tener valores erróneos al realizar el resúmen estadístico de las variables númericas.

```python
df['Diverted'] = df['Diverted'].astype('category')
df['FlightNum'] = df['FlightNum'].astype('category')
```

```python
df.describe().applymap("{0:.2f}".format)
```

Creamos la nueva variable "Delayed" (Retraso)

```python
#Añadimos la columna Delayed,si ha llegado tarde (True) y sino (False)
df['Delayed']=df.ActualElapsedTime>df.CRSElapsedTime

df[['ActualElapsedTime','CRSElapsedTime','Delayed']].head(10)
```

Creamos la variable "veloc_mean" (Velocidad Media)

```python
df['veloc_mean']=round(df['Distance']*60/df['AirTime'],2)

df.groupby('UniqueCarrier')[['UniqueCarrier','veloc_mean']].head()
```

¿Cuáles son los vuelos más largos? 

```python
vuelos_largos = df[['FlightNum','Origin','Dest','Distance']].groupby(['Distance']).first().reset_index()
vuelos_largos.sort_values(by=['Distance'],ascending=False)[:10]
```

¿Y los más atrasados?

```python
#Atraso en la llegada:
df[['FlightNum','UniqueCarrier','Origin','Dest','ArrDelay']].sort_values(by='ArrDelay',ascending=False)[:10]
```

```python
#Atraso en la salida:
df[['FlightNum','UniqueCarrier','Origin','Dest','DepDelay']].sort_values(by='DepDelay',ascending=False)[:10]
```

Tabla de los aeropuertos con mas retrasos acumulados

```python
df.groupby('UniqueCarrier').sum()['Delayed'].sort_values(ascending=False).head(10)
```
